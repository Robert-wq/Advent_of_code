```python
df = pd.read_csv('input.txt', names=['number'], dtype=str)
df
```

```python
num_arr = df['number'].to_numpy()
num_arr
```

- 12 times
- Loop through each binary value 
- count how many ones and zeros there are
- the most common is added to the gamma total
- the least common is added to the epsilon total.
- multiply the two values

```python
gamma = 0
epsilon = 0

for i in range(len(num_arr[0])):
    one_count = 0
    zero_count = 0
    for j in range(len(num_arr)):
        if num_arr[j][i] == '1':
            one_count += 1
        else:
            zero_count += 1
            
    if(one_count >= zero_count):
        #gets the number from 2 ** 11 to 2 ** 0, had to add the extra '-1' to account for 0 index
        gamma += (2 ** ((len(num_arr[0]) - 1) - i)) 
    else:
        epsilon += (2 ** ((len(num_arr[0]) - 1) - i))
    
    print(str(len(num_arr[0]) - i) + ' = ' + str(one_count))
    print('gamma = ' + str(gamma))
    print('epsilon = ' + str(epsilon))
```

```python
result = gamma * epsilon
result
```
